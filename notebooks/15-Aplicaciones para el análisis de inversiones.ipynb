Debemos comenzar distinguiendo la tasa de interes nominal y la tasa de interes efectiva. Esto es con el objetivo de obtener la tasa real de nuestra inversion.

* ***Tasa nominal*** = `es una tasa de interés  que no toma en cuenta la frecuencia con la que se capitalizan los intereses `(es decir, si se reinvierten o no). Es decir, la nominal refiere a las normas o presentaciones.

* ***Tasa efectiva*** = `sí considera la capitalización de los intereses y muestra el costo o rendimiento real de un producto financiero en un período determinado, ya sea un préstamo o una inversión`. Es decir, cuanto obtengo de intereses absolutos (es lo que nos interesa.)

### **1) Formula de equivalencia**
¿Como pasamos de un periodo a a otro?🧮

La fórmula general para convertir por ejemplo una TEA (Tasa Efectiva Anual) a TES (Tasa Efectiva Semestral) es:

$$i(t_2) = \left(1 + i(t_1)\right)^{\frac{t_2}{t_1}} - 1$$


### 📌 Donde:
- $it2$ es la **tasa nominal buscada** (en decimales, por ejemplo: 12% = 0.12).
- $t1$ es **el periodo que tengo como dato**
- $t2$ es **el periodo que tengo que buscar**

Pasamos un dato de un periodo a otro periodo


---







### 🔑 Nominal vs Efectiva, en criollo

Pensá en la tasa **nominal** como el **precio de cartel**: el número que el banco te muestra en grande (*"¡40% anual!"*). Es cómodo para comunicar, pero **no te dice cuánto ganás/pagás de verdad**, porque ignora cada cuánto se suman los intereses.

La tasa **efectiva** es **lo que realmente te queda**: ya tiene en cuenta que los intereses se van sumando y empiezan a generar más intereses.

🚗 *Analogía:* la nominal es como decir *"voy a 120 km/h"*. La efectiva es **cuántos kilómetros recorriste de verdad**, que depende de cuánto tiempo mantuviste esa velocidad. El cartel dice una cosa; el camino real, otra.

> **🧭 Regla de oro para no marearte:**
> - Si la cuenta **multiplica** la tasa (× días, × meses) → es **nominal / simple**, *NO capitaliza*.
> - Si la cuenta **eleva a una potencia** $(1+i)^n$ → es **efectiva / compuesta**, *SÍ capitaliza*.
>
> Mirá el exponente: **si hay potencia, hay capitalización.**

## 🏔️ ¿Qué es "capitalizar"? Una bola de nieve

**Capitalizar = los intereses que ganás se suman al capital, y en el período siguiente esos intereses también generan intereses.** Es *interés sobre interés*: una bola de nieve que arranca chica y se agranda sola.

Veámoslo con $1.000 al 10% anual durante 3 años:

| Año | Empiezo con | Gano (10%) | Termino con |
|-----|-------------|------------|-------------|
| 1   | \$1.000     | \$100      | \$1.100     |
| 2   | \$1.100     | **\$110**  | \$1.210     |
| 3   | \$1.210     | **\$121**  | \$1.331     |

Fijate que **cada año ganás más que el anterior**, aunque la tasa sea siempre 10%. ¿Por qué? Porque el segundo año cobrás interés también sobre los \$100 del primer año.

- **Sin capitalizar** (interés simple): \$100 + \$100 + \$100 → llego a **\$1.300**.
- **Capitalizando** (interés compuesto): llego a **\$1.331**.

Esos **\$31 de más** son *"el interés que generó el interés"*. Parece poco con 3 años y 10%… pero con **tasas altas o muchos períodos esta diferencia se dispara** (lo vamos a ver clarísimo con la inflación 👇).

Por eso la fórmula del valor futuro usa una **potencia**: el exponente $n$ es **cuántas veces reinvertís**. Cada $(1+i)$ que multiplicás arrastra TODO lo acumulado, no solo el capital original.

### ***2) Formula de Proporcion***
¿Como pasamos de EFECTIVA a NOMINAL?🧮

La fórmula general para convertir una TN a TE es:

$$
TNA = i_{\text{efectiva, periodo}} \cdot \frac{365}{n}
$$

$$
i_{\text{efectiva, periodo}}  = TNA \cdot \frac{n}{365}
$$

### 📌 Donde:

- $TNA$ → Tasa nominal anual  
- $i_efectiva$ , periodo → Tasa efectiva para un período de \(n\) días  
- $ n $ → Número de días del período de la tasa efectiva

Ojo: si la tasa efectiva del período es compuesta y querés obtener la nominal equivalente, hay que usar equivalencia primero y no solo multiplicar por

## ***3) Valor futuro***

La fórmula:

$$
C_n = C_0 \cdot (1 + i)^n
$$

describe el **régimen de capitalización compuesta**. Es decir, cómo crece una inversión o deuda a lo largo del tiempo cuando los intereses se reinvierten al final de cada período.

### 🔎 Desglose:

- $ C_0 $: capital inicial, es decir, lo que invertís o prestás hoy.
- $ i $: tasa de interés por período (por ejemplo, mensual si capitalizás mensualmente).
- $ n $: número total de períodos (por ejemplo, meses).
- $C_n $: capital final acumulado después de \( n \) períodos, con interés compuesto.

---






## ***4) Valor Presente***

La fórmula:

$$
C_0 = \frac{C_n}{(1 + i)^n}
$$

Es básicamente la inversa del cálculo de valor futuro: en lugar de proyectar cuánto tendrás más adelante, traés un monto futuro a valor de hoy. Esto se llama descontar o actualizar un valor.



### 🔎 Desglose:
- C_0: Capital inicial o valor presente, lo que invertís o prestás hoy.
- C_n: Capital final o valor futuro, monto acumulado al final.
- i: Tasa de interés por período (por ejemplo, mensual si capitalizás mensualmente).
- n: Número total de períodos (meses, años, etc.).

# Formulas aplicadas a la vida real ⏰



Comencemos definiendo todas las formulas que hemos visto hasta el momento

```python
# 1) Fórmula de Proporción
def proporcion(TNA, n):
    return TNA * (n / 365)

# 2) Fórmula de Equivalencia
def equivalencia(i_t1, t2, t1):
    return (1 + i_t1) ** (t2 / t1) - 1

# 3) Capitalización Compuesta
def capitalizacion_compuesta(C0, i, n):
    return C0 * (1 + i) ** n

# 4) Valor presente
def valor_presente(Cn, i, n):
    return Cn/((1 + i) ** n )
```

### 🧰 Machete: ¿qué fórmula uso y para qué?

Ya tenemos las 4 funciones cargadas. Esta tablita es tu mapa para saber cuál agarrar:

| Quiero… | Uso | En criollo | ¿Capitaliza? |
|---------|-----|-----------|:---:|
| Repartir una TNA a un plazo más corto (la cuenta "rápida" del banco) | `proporcion(TNA, n)` | Parto la tasa anual en pedacitos lineales | ❌ |
| Pasar una tasa efectiva de un plazo a otro (mensual↔anual…) | `equivalencia(i, t2, t1)` | "Estiro" o "achico" la tasa respetando el interés compuesto | ✅ |
| Saber en cuánto se convierte mi plata en el futuro | `capitalizacion_compuesta(C0, i, n)` | Llevo plata **hacia adelante** en el tiempo | ✅ |
| Saber cuánto vale **hoy** una plata que voy a cobrar después | `valor_presente(Cn, i, n)` | Traigo plata del futuro a hoy (**descontar**) | ✅ |

💡 **Equivalencia vs Proporción** es la pareja que más confunde:
- **Proporción** reparte la tasa de forma lineal (multiplicando). Es rápida y aproximada. La usás cuando te dan una TNA y querés el pedacito de un subperíodo.
- **Equivalencia** convierte respetando la capitalización (potencia). Es la "correcta" cuando lo que ya tenés es una tasa efectiva y la querés en otro plazo.

Si al mismo problema le aplicás las dos, **dan distinto** — y esa diferencia es, otra vez, la **capitalización**.

ejemplo: Santander ofrece creditos a una tasa nominal anual de 40%. ¿cuanto seria efectivamente la tasa a los 90 dias?

## 🧪 Comprobémoslo con código: ¿por qué proporción y equivalencia son distintas?

Ya dijimos que **proporción NO capitaliza** (reparte lineal) y **equivalencia SÍ capitaliza** (interés sobre interés). Para no solo "creernos la teoría", vamos a *demostrarlo* con tres mini-experimentos. 👇

### 🟢 Experimento 1 — La proporción es "lineal": no importa el plazo, el total es el mismo

Tomemos una **TNA del 36%** y \$1.000. Vamos a repartir esa tasa anual en pedacitos (mensual, trimestral, semestral) con `proporcion`, y después acumular el interés de forma **simple** durante todo el año.

La pregunta: ¿da lo mismo medir la tasa por mes que por trimestre? 🤔

```python
TNA, C0 = 0.36, 1000

print("Reparto la TNA del 36% en distintos plazos y acumulo SIMPLE todo el año:\n")

# Mensual (30 días)
dias = 30
i_mensual = proporcion(TNA, dias)
periodos = 365 / dias
interes = i_mensual * periodos
monto = C0 * (1 + interes)
print(f"  {'mensual':11s}: tasa del período = {i_mensual*100:6.3f}%  ->  total año = {interes*100:.2f}%  ->  ${monto:,.2f}")

# Trimestral (90 días)
dias = 90
i_trim = proporcion(TNA, dias)
periodos = 365 / dias
interes = i_trim * periodos
monto = C0 * (1 + interes)
print(f"  {'trimestral':11s}: tasa del período = {i_trim*100:6.3f}%  ->  total año = {interes*100:.2f}%  ->  ${monto:,.2f}")

# Semestral (180 días)
dias = 180
i_sem = proporcion(TNA, dias)
periodos = 365 / dias
interes = i_sem * periodos
monto = C0 * (1 + interes)
print(f"  {'semestral':11s}: tasa del período = {i_sem*100:6.3f}%  ->  total año = {interes*100:.2f}%  ->  ${monto:,.2f}")
```

**¿Qué pasó?** Los tres caminos dan **exactamente \$1.360**. 🎯

Esto es lo lindo de la proporción: como solo **multiplica y divide** por el plazo, esas operaciones se cancelan entre sí. Repartir la tasa en pedacitos chiquitos o grandes da igual, *siempre que no capitalices*. 

> 💡 Por eso la proporción es la cuenta "rápida" del banco: es simple y predecible pero **ignora** que en la vida real los intereses muchas veces se reinvierten.

### 🔵 Experimento 2 — "Dos caminos, mismo destino": las tasas equivalentes preservan el monto

Ahora al revés. Tengo una tasa **efectiva anual del 60%** y quiero invertir \$1.000 por un año. Lo puedo hacer:
- **Directo:** capitalizo una vez al 60%.
- **Dando vueltas:** convierto el 60% anual a una tasa mensual (o trimestral, o semestral) **equivalente** con `equivalencia`, y después capitalizo ese pedacito las veces que haga falta.

Si la equivalencia está bien hecha, **todos los caminos tienen que llegar al mismo monto final.** Probémoslo:

```python
TEA, C0 = 0.60, 1000

print(f"Camino directo (1 año al 60%):           ${C0*(1+TEA):,.2f}\n")
print("Caminos 'dando vueltas' (convierto y capitalizo):")

# Mensual (30 días)
dias = 30
i_mensual = equivalencia(TEA, dias, 365)
monto = capitalizacion_compuesta(C0, i_mensual, 365/dias)
print(f"  vía tasa {'mensual':11s} ({i_mensual*100:6.3f}%):  ${monto:,.2f}")

# Trimestral (90 días)
dias = 90
i_trim = equivalencia(TEA, dias, 365)
monto = capitalizacion_compuesta(C0, i_trim, 365/dias)
print(f"  vía tasa {'trimestral':11s} ({i_trim*100:6.3f}%):  ${monto:,.2f}")

# Semestral (180 días)
dias = 180
i_sem = equivalencia(TEA, dias, 365)
monto = capitalizacion_compuesta(C0, i_sem, 365/dias)
print(f"  vía tasa {'semestral':11s} ({i_sem*100:6.3f}%):  ${monto:,.2f}")
```

**¿Qué pasó?** Todos los caminos llegan a **\$1.600**. 

No importa si medís la tasa por mes, por trimestre o por semestre: si son **tasas equivalentes**, el monto final es idéntico. Eso es lo que significa "equivalentes" → distintas formas de expresar **el mismo rendimiento real**.

> 💡 Compará con el experimento anterior: la proporción daba \$1.360 y la equivalencia da \$1.600 sobre tasas parecidas. Esos **\$240 de diferencia son la capitalización**: el interés que generó interés.

### ⚖️ Experimento 3 — ¿Cuánto importa capitalizar? Mirá cómo se abre la brecha

Con poco tiempo, simple y compuesto casi no se diferencian. Pero con muchos períodos, la bola de nieve se vuelve gigante. Tomemos \$1.000 al **5% por período** y veamos qué pasa a medida que pasa el tiempo:

```python
import matplotlib.pyplot as plt
import numpy as np

C0, i = 1000, 0.05

print(f"{'Períodos':>9} | {'Simple':>12} | {'Compuesto':>12} | {'Diferencia':>12}")
print("-"*54)

for n in [1, 5, 10, 20, 40]:
    simple = C0 * (1 + i*n)
    compuesto = C0 * (1 + i)**n
    print(f"{n:>9} | ${simple:>10,.2f} | ${compuesto:>10,.2f} | ${compuesto-simple:>10,.2f}")

# ===== Gráfico comparativo =====
n_vals = np.arange(1, 40)
simple_vals = C0 * (1 + i * n_vals)
comp_vals = C0 * (1 + i) ** n_vals

plt.figure()
plt.plot(n_vals, simple_vals, label="Interés simple")
plt.plot(n_vals, comp_vals, label="Interés compuesto")

plt.xlabel("Períodos (n)")
plt.ylabel("Monto ($)")
plt.title("Interés simple vs compuesto")
plt.legend()
plt.show()
```

**La moraleja:** en el período 1 son idénticos, pero al período 40 el compuesto **más que duplica** al simple. 

Por eso usar la fórmula correcta importa MUCHO cuando hay tasas altas o plazos largos — exactamente el contexto argentino que vemos abajo con la inflación. Elegir mal entre proporción y equivalencia puede cambiar por completo el número que reportás.

```python
# Inflación mayorista MENSUAL (en decimales)
i_dic23 = 0.54   # dic-2023, mes de la devaluación
i_2024  = 0.012   

# Anualizar una tasa MENSUAL = capitalizarlo 12 veces
def anualizar_mensual(i_mensual):
    return (1 + i_mensual) ** 12 - 1

anual_dic23 = anualizar_mensual(i_dic23)
anual_2024  = anualizar_mensual(i_2024)

print(f"Mensual dic-23: {i_dic23*100:.1f}%  ->  ritmo anualizado: {anual_dic23*100:,.1f}%")
print(f"Mensual  2024 : {i_2024*100:.1f}%   ->  ritmo anualizado: {anual_2024*100:,.1f}%")
print(f"\nLe pegó justo!")
```

Importante: 'anualizar' un mes capitaliza ese ritmo 12 veces, así que responde a la pregunta 'si este mes se repitiera todo el año, ¿cuánto daría?'. No es la inflación efectivamente acumulada del año, sino una proyección del ritmo. La comparación muestra el fuerte descenso del ritmo de inflación, _si el ritmo se mantuviera_.

### 🤔 último tema: ¿prefieren \$10.000 hoy o \$15.000 en 3 años?

Te ponen sobre la mesa dos opciones:
- **Opción A:** \$10.000 **hoy**, en la mano. 💵
- **Opción B:** \$15.000 pero dentro de **3 años**. ⏳

A simple vista \$15.000 > \$10.000, así que parecería ganar la B… pero ya sabemos que **comparar plata de distintos momentos sin descontar es hacer trampa**. Para decidir bien, traemos los \$15.000 del futuro a valor de hoy con `valor_presente` y recién ahí comparamos peras con peras.

¿Y qué tasa uso para descontar? La que represente **tu mejor alternativa** (lo que podrías ganar con esos \$10.000 si los tuvieras hoy). Probemos varias 👇

```python
hoy, futuro, n = 10000, 15000, 3

print("Comparo trayendo los $15.000 a valor de HOY:\n")
for tasa in [0.05, 0.10, 0.20]:
    vp = valor_presente(futuro, tasa, n)
    decision = "conviene ESPERAR (Opción B)" if vp > hoy else "conviene AGARRAR HOY (Opción A)"
    print(f"  Si puedo ganar {tasa*100:2.0f}% anual -> los $15.000 valen hoy ${vp:,.2f}  ->  {decision}")
```

**¡La decisión depende de la tasa!** 🤯
- Con tasas **bajas** (5%, 10%), esperar conviene: descontar poco deja a los \$15.000 valiendo más que \$10.000 hoy.
- Con tasas **altas** (20%), conviene agarrar hoy: si podés hacer rendir tu plata al 20%, \$10.000 hoy se convierten en más de \$15.000 en 3 años por tu cuenta.

> _BONUS:_ ¿Existe una tasa donde **dé exactamente igual** una opción que la otra? Sí: es la **tasa de indiferencia** (o *break-even*). La despejamos pidiendo que el valor presente del futuro iguale a los \$10.000 de hoy:

```python
# Tasa que hace VP(futuro) = hoy  ->  hoy = futuro / (1+i)^n  -> despejando ->  i = (futuro/hoy)^(1/n) - 1
tasa_indiferencia = (futuro/hoy)**(1/n) - 1
print(f"Tasa de indiferencia: {tasa_indiferencia*100:.2f}%")
print(f"Comprobación -> VP de los $15.000 a esa tasa: ${valor_presente(futuro, tasa_indiferencia, n):,.2f}")
```

>**Lectura final** 🧭: la tasa de indiferencia es **14,47%**. Es tu vara de decisión:
>- Si tu mejor alternativa rinde **menos** del 14,47% → esperá los \$15.000.
>- Si rinde **más** → agarrá los \$10.000 hoy.
>
>Y esta es justamente la lógica del **VAN** que vemos a continuación: descontar flujos futuros a una tasa y compararlos contra lo que ponés hoy. El VAN no es más que **este mismo razonamiento, >
>pero con varios flujos a la vez**. 🚀

## 5) Valor Actual Neto (VAN)

El Valor Actual Neto (VAN) es uno de los indicadores más importantes para evaluar la viabilidad económica de un proyecto de inversión. `Se calcula trayendo a valor presente todos los flujos de caja futuros esperados del proyecto y restándole la inversión inicial.`

La fórmula general del VAN es:

$$
VAN = \sum_{t=0}^{n} \frac{F_t}{(1+r)^t} - I_0
$$

Donde:

- $ VAN $: Valor Actual Neto
- $ F_t $: Flujo de caja en el período \( t \)
- $ r $: Tasa de descuento 
- $ t $: Período de tiempo
- $ n $: Número total de períodos
- $ I_0 $: Inversión inicial (que generalmente se representa como un flujo de caja negativo en el período 0)

**Interpretación del VAN:**

- Si **VAN > 0**: El proyecto es rentable y se espera que genere valor por encima de la tasa de rendimiento requerida. Conviene aceptar el proyecto.
- Si **VAN < 0**: El proyecto no es rentable y se espera que destruya valor. No conviene aceptar el proyecto.
- Si **VAN = 0**: El proyecto es indiferente. Genera exactamente la tasa de rendimiento requerida, pero no crea valor adicional.


# ¿Cual es la diferencia con el valor presente?

* El valor presente (VP) es el valor actual de uno o varios flujos futuros descontados a una tasa determinada, mientras que el valor actual neto (VAN) es ese valor presente total menos la inversión inicial. En otras palabras, el VP te dice cuánto valen hoy los flujos futuros, y el VAN te indica si el proyecto crea o destruye valor

---


### 💡 El VAN en criollo

Imaginá que alguien te propone un negocio: ponés plata hoy y te promete cobros durante varios años. La pregunta del millón es: **¿me meto o no?**

El VAN lo resuelve en 3 pasos:
1. **Traigo cada cobro futuro a valor de hoy** (lo descuento, porque ya sabemos que la plata del futuro vale menos).
2. **Sumo todos esos valores de hoy.**
3. **Le resto lo que tengo que invertir hoy.**

- Si el resultado es **positivo** ➕ → el proyecto te deja **más** de lo que pedías como mínimo → **conviene**.
- Si es **negativo** ➖ → te deja menos → **no conviene**.

🔑 La **tasa de descuento** $r$ es tu **vara de exigencia**: el rendimiento mínimo que le pedís a la plata (lo que ganarías en tu mejor alternativa). Cuanto más alta la vara, más difícil que el proyecto la pase. Por eso mover esa tasa **puede dar vuelta la decisión** — lo vas a poder probar vos mismo con el slider de más abajo. 🎚️


Ahora, veamos un ejemplo práctico en código.

```python
#lista
flujos_caja = {
    'Año': [0, 1, 2, 3, 4, 5],
    'Flujo': [-10000, 3000, 3500, 4000, 4000, 2000]
}

# Creo df
df_flujos = pd.DataFrame(flujos_caja)

df_flujos
```

```python
# Definir la tasa de descuento para ver cuanto saca periodo a periodo
tasa_descuento = 0.50

# valor presente
df_flujos['Valor Presente'] = df_flujos['Flujo'] / (1 + tasa_descuento) ** df_flujos['Año'] # por cada flujo lo descuenta por una tasa de interes

# Calcular el VAN sumando los valores presentes de todos los flujos
van_calculado = df_flujos['Valor Presente'].sum()

# Mostrar el DataFrame con los valores presentes
print("Tabla de Flujos de Caja y sus Valores Presentes:")
display(df_flujos)

# Mostrar el resultado del VAN
print(f"\nEl Valor Actual Neto (VAN) del proyecto es: ${van_calculado:.2f}")
```

```python
# Interpretar el resultado del VAN
if van_calculado > 0:
    print("Interpretación: El VAN es positivo, por lo tanto, el proyecto es rentable y conviene aceptarlo.")
elif van_calculado < 0:
    print("Interpretación: El VAN es negativo, por lo tanto, el proyecto no es rentable y no conviene aceptarlo.")
else:
    print("Interpretación: El VAN es cero, el proyecto es indiferente.")
```

generar una formula para que se aplique para otros casos y despues hacer un ejemplo con un slide

```python
#Guardemos una funcion donde calcule VPN
def calcular_vpn(flujos, tasa):
    ##Formula que calcula VPN, toma un array y una tasa la cual debemos definir con ese mismo nombre)##

    df = pd.DataFrame(flujos)
    df['VP'] = df['Flujo'] / (1 + tasa) ** df['Año']
    vpn = df['VP'].sum()
    return df, vpn
```

```python
# Inputs del usuario
# Usar inputs manuales si se desea

# Flujos predefinidos por ahora
flujos = {
    'Año': [0, 1, 2, 3, 4],
    'Flujo': [-5000, 1500, 2000, 2000, 1000]
}
tasa = 0.1

df, vpn = calcular_vpn(flujos, tasa)
print(df)
print(f"\n📌 Valor Presente Neto del proyecto: ${vpn:.2f}")
```

```python
plt.bar(df['Año'] - 0.2, df['Flujo'], width=0.4, label='Flujo', align='center')
plt.bar(df['Año'] + 0.2, df['VP'], width=0.4, label='Valor Presente', align='center')
plt.xlabel('Año')
plt.ylabel('Monto ($)')
plt.title('Flujos y Valores Presentes')
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Sliders para tasa y años

tasa_slider = widgets.FloatSlider(
    value=0.1,
    min=0.01,
    max=0.5,
    step=0.01,
    description='Tasa:',
    readout_format='.2%',
    continuous_update=False
)

anios_slider = widgets.IntSlider(
    value=4,
    min=1,
    max=10,
    step=1,
    description='Años:',
    continuous_update=False
)
```

```python
def interfaz_vpn(tasa, anios):
    np.random.seed(42) # seteamos una semilla aleatoria
    flujos = [-5000] + list(np.random.randint(800, 2500, size=anios)) #hacemos un flujo de dinero que comience negativo producto de la inversion inicial y que luego vaya asignando de manera aleatoria valores entre 800 & 2500
    años = list(range(anios + 1)) # aramar años-flujo en funcion del rango que tome el tome el slider

    datos = {'Año': años, 'Flujo': flujos}
    df, vpn = calcular_vpn(datos, tasa) #VPN

    clear_output(wait=True)
    display(tasa_slider, anios_slider)

    print(f"\n📌 Valor Presente Neto (VPN): ${vpn:.2f}")
    print(df)


    plt.bar(df['Año'] - 0.2, df['Flujo'], width=0.4, label='Flujo', align='center') #PLOT
    plt.bar(df['Año'] + 0.2, df['VP'], width=0.4, label='Valor Presente', align='center')
    plt.xlabel('Año')
    plt.ylabel('Monto ($)')
    plt.title('Flujos y Valores Presentes')
    plt.axhline(0, color='black', linewidth=0.8)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

widgets.interact(interfaz_vpn, tasa=tasa_slider, anios=anios_slider);
```
